### 1. A car with a full tank of gas drives 80 miles per hour for 3 hours before running out of gas. How many miles can the car go per tank?
To solve this problem, we use the formula $${S} = \frac {D} {T} $$ <br>
&emsp;Where,S = speed in miles per hour <br>
&emsp;&emsp;&emsp;&emsp;D = distance in miles <br>
&emsp;&emsp;&emsp;&emsp;T = time in hours <br>
We can now cross multiply to solve for D, we now have $ \mathbf S X T $

```python
80*3
```

#### Therefore, the car can go 240 miles per tank.

### 2. A runner travels a half marathon (13.1 miles) in 3 hours. How fast was this runner going in miles per hour?
To solve this problem, we use the same method in question (1) above, but this time we are solving for S which is $ \frac {D} {T} $

```python
13.1/3
```

>The runner was going 4.36 miles per hour.

### 3. The spread of a virus is exponential. If 5 people are infected and each person infects 2 other people, how many people have been infected after 5 rounds of infection?
Formula for exponential infections is $ \mathbf y=ab^x $ <br>
where a=initial number of infected bacteria <br>
&emsp;&emsp;&emsp; b=number infected by one infected bacteria <br>
&emsp;&emsp;&emsp; x=number of infection rounds

```python
5*(2**5)
```

*160 people were infected after 5 rounds of infection.*
